```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
from math import log, pow, exp
```

```python
def rand(error):
    L = 10000
    return (rd.random() * rd.randint(-L, L)) % error

def func_mnk(vect, x):
    res = 0
    for i in range(len(vect) - 1, -1, -1):
        res += vect[i] * (x ** (len(vect) - 1 - i))
    return res

def func_mnk_exp(vect, x):
    return exp(vect[0]) * (x ** vect[1])
```

## random function

```python
N = 5
arr_x = [float(i+0.01) for i in range(N)]
arr_y = [rand(2) for _ in range(N)]

#arr_x = [2, 3, 4, 5, 6]
#arr_y = [1, 4, 1, 5, 2]

#linear function
matr = np.array([ 
    [sum(arr_x)] + [N],
    [sum(map(lambda x: x*x, arr_x))] + [sum(arr_x)] ])
val = np.array([sum(arr_y)] + [sum(arr_x[i] * arr_y[i] for i in range(N))]) 
vect = np.linalg.solve(matr, val)

step = 0.1
arr_x_mnk = np.arange(min(arr_x), max(arr_x) + step, step)

plt.plot(arr_x_mnk, [func_mnk(vect, x) for x in arr_x_mnk], c="red")
plt.plot(arr_x, arr_y, 'x', c=[0, 0, 0])

#quadratic function

matr = np.array([ 
    [sum(map(lambda x: x**2, arr_x))] + [sum(map(lambda x: x**1, arr_x))] + [sum(map(lambda x: x**0, arr_x))],
    [sum(map(lambda x: x**3, arr_x))] + [sum(map(lambda x: x**2, arr_x))] + [sum(map(lambda x: x**1, arr_x))],
    [sum(map(lambda x: x**4, arr_x))] + [sum(map(lambda x: x**3, arr_x))] + [sum(map(lambda x: x**2, arr_x))]
    ])
val = np.array(
    [sum(arr_y[i] * (arr_x[i]**0) for i in range(N))] + 
    [sum(arr_y[i] * (arr_x[i]**1) for i in range(N))] +
    [sum(arr_y[i] * (arr_x[i]**2) for i in range(N))]
    ) 
vect = np.linalg.solve(matr, val)

plt.plot(arr_x_mnk, [func_mnk(vect, x) for x in arr_x_mnk],c="blue")

#exp function

matr = np.array([   
    [N] + [sum(map(log, arr_x))],
    [sum(map(log, arr_x))] + [sum(map(lambda x:x*x, list(map(log, arr_x))))]
    ])
val = np.array(
    [sum(map(log, arr_y))] + 
    [sum(list(map(log, arr_x))[i] * list(map(log, arr_y))[i] for i in range(N))]
    )
vect = np.linalg.solve(matr, val)

plt.plot(arr_x_mnk, [func_mnk_exp(vect, x) for x in arr_x_mnk],c="green")

plt.show()
```

## f(x) = a * x^b

```python
N = 10
arr_x = [float(i+0.01) for i in range(N)]
arr_y = []

a = rd.randint(1, 1000) / 1000
b = rd.randint(1, 1000) / 1000

for x in arr_x:
    arr_y.append(a * x ** b)

#linear function
matr = np.array([ 
    [sum(arr_x)] + [N],
    [sum(map(lambda x: x*x, arr_x))] + [sum(arr_x)] ])
val = np.array([sum(arr_y)] + [sum(arr_x[i] * arr_y[i] for i in range(N))]) 
vect = np.linalg.solve(matr, val)

step = 0.1
arr_x_mnk = np.arange(min(arr_x), max(arr_x) + step, step)

plt.plot(arr_x_mnk, [func_mnk(vect, x) for x in arr_x_mnk], c="red")
plt.plot(arr_x, arr_y, 'x', c=[0, 0, 0])

#quadratic function

matr = np.array([ 
    [sum(map(lambda x: x**2, arr_x))] + [sum(map(lambda x: x**1, arr_x))] + [sum(map(lambda x: x**0, arr_x))],
    [sum(map(lambda x: x**3, arr_x))] + [sum(map(lambda x: x**2, arr_x))] + [sum(map(lambda x: x**1, arr_x))],
    [sum(map(lambda x: x**4, arr_x))] + [sum(map(lambda x: x**3, arr_x))] + [sum(map(lambda x: x**2, arr_x))]
    ])
val = np.array(
    [sum(arr_y[i] * (arr_x[i]**0) for i in range(N))] + 
    [sum(arr_y[i] * (arr_x[i]**1) for i in range(N))] +
    [sum(arr_y[i] * (arr_x[i]**2) for i in range(N))]
    ) 
vect = np.linalg.solve(matr, val)

plt.plot(arr_x_mnk, [func_mnk(vect, x) for x in arr_x_mnk],c="blue")

#exp function

matr = np.array([   
    [N] + [sum(map(log, arr_x))],
    [sum(map(log, arr_x))] + [sum(map(lambda x:x*x, list(map(log, arr_x))))]
    ])
val = np.array(
    [sum(map(log, arr_y))] + 
    [sum(list(map(log, arr_x))[i] * list(map(log, arr_y))[i] for i in range(N))]
    )
vect = np.linalg.solve(matr, val)

plt.plot(arr_x_mnk, [func_mnk_exp(vect, x) for x in arr_x_mnk],c="green")

plt.show()
```
